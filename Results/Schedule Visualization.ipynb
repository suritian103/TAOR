```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter

def generate_schedules(course_csv_path, workshop_csv_path, course_names, room_names):
    # 读取课程CSV文件，并只保留标记为1的行
    df_courses = pd.read_csv(course_csv_path)
    df_courses = df_courses[df_courses.iloc[:, -1] == 1]
    df_courses['rooms'] = df_courses['rooms'].apply(lambda x: room_names[x-1])
    df_courses['courses'] = df_courses['courses'].apply(lambda x: course_names[x-1])
    
    # 读取Workshop CSV文件，并只保留标记为1的行
    df_workshops = pd.read_csv(workshop_csv_path)
    df_workshops = df_workshops[df_workshops.iloc[:, -1] == 1]
    df_workshops['rooms'] = df_workshops['rooms'].apply(lambda x: room_names[x-1])
    df_workshops['courses'] = df_workshops.apply(lambda row: f"{course_names[row['courses']-1]} (Group {row['groups']})", axis=1)
    
    # 合并过滤后的课程和Workshop信息
    df_combined = pd.concat([df_courses.drop(df_courses.columns[-1], axis=1), df_workshops.drop(df_workshops.columns[-1], axis=1)], ignore_index=True)
    
    # 为每个房间生成课程表
    for room in df_combined['rooms'].unique():
        room_df = df_combined[df_combined['rooms'] == room]
        
        # 初始化空的DataFrame来存储当前房间的课程安排
        room_schedule = pd.DataFrame(index=pd.MultiIndex.from_product([room_df['weeks'].unique(), room_df['days'].unique()],
                                                                       names=['Week', 'Day']),
                                      columns=room_df['slots'].unique()).sort_index().sort_index(axis=1)
        
        # 填充课程表
        for _, row in room_df.iterrows():
            current_slot = room_schedule.at[(row['weeks'], row['days']), row['slots']]
            new_entry = row['courses']
            if pd.isna(current_slot):
                room_schedule.at[(row['weeks'], row['days']), row['slots']] = new_entry
            else:
                room_schedule.at[(row['weeks'], row['days']), row['slots']] = f"{current_slot}; {new_entry}"
        
        # 生成Excel文件名
        excel_file_name = f'{room}_schedule_filtered.xlsx'
        
        # 创建一个新的Excel工作簿和工作表
        wb = Workbook()
        ws = wb.active
        ws.title = "Schedule"
        
        # 将DataFrame转换为工作表中的行
        for r_idx, row in enumerate(dataframe_to_rows(room_schedule, index=True, header=True), 1):
            for c_idx, value in enumerate(row, 1):
                ws.cell(row=r_idx, column=c_idx, value=value)
        
        # 调整列宽
        for column_cells in ws.columns:
            length = max(len(str(cell.value)) for cell in column_cells)
            ws.column_dimensions[get_column_letter(column_cells[0].column)].width = length
        
        # 保存工作簿
        wb.save(filename=excel_file_name)
        
        print(f'Saving schedul of {room}：{excel_file_name}')
# 用法示例（请替换路径和列表为实际值）
# generate_schedules_with_filtered_sessions('path/to/course.csv', 'path/to/workshop.csv', course_names, room_names)



# 使用函数生成带有课程名称的课程表
course_names = [
    "Accelerated Algebra and Calculus for Direct Entry",
    "Accelerated Proofs and Problem Solving",
    "Analytic Number Theory",
    "Applied Stochastic Differential Equations",
    "Axiomatic Set Theory",
    "Bayesian Theory",
    "Biostatistics",
    "Combinatorics and Graph Theory",
    "Commutative Algebra",
    "Differential Geometry",
    "Discrete-Time Finance",
    "Engineering Mathematics 1a",
    "Entrepreneurship in the Mathematical Sciences",
    "Essentials in Analysis and Probability",
    "Facets of Mathematics",
    "Finance, Risk and Uncertainty",
    "Fundamentals of Operational Research",
    "Fundamentals of Optimization",
    "General Topology",
    "Generalised Regression Models",
    "Geometry",
    "Group Theory",
    "Honours Analysis",
    "Honours Differential Equations",
    "Industrial Mathematics",
    "Introduction to Data Science",
    "Introduction to Lie Groups",
    "Introduction to Linear Algebra",
    "Introduction to Linear Algebra",  # 重复项
    "Introduction to Partial Differential Equations",
    "Introductory Mathematics with Applications",
    "Linear Analysis",
    "Mathematical Biology",
    "Mathematics for the Natural Sciences 1a",
    "Methodology, Modelling and Consulting Skills",
    "Numerical Linear Algebra",
    "Probability",
    "Python Programming",
    "Several Variable Calculus and Differential Equations",
    "Several Variable Calculus and Differential Equations",  # 重复项
    "Statistical Methodology",
    "Statistical Programming",
    "Stochastic Analysis in Finance",
    "Stochastic Modelling",
    "Theory of Statistical Inference",
    "Topics in Ring and Representation Theory",
    "Variational Calculus"
]

room_names = [
    "JCMB_1501", "JCMB_2901", "JCMB_5326", "JCMB_5327", "JCMB_5328", "JCMB_6206",
    "JCMB_Lecture Theatre A", "JCMB_Lecture Theatre B", "JCMB_Lecture Theatre C",
    "NUC_1.14 - Oak Lecture Theatre", "NUC_B.01 - Alder Lecture Theatre", "NUC_1.15 - Larch Lecture Theatre"
]

generate_schedules('Course_S1.csv', 'Workshop_S1_update.csv',course_names, room_names)
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter

def generate_schedules(course_csv_path, workshop_csv_path, course_names, room_names):
    # 读取课程CSV文件，并只保留标记为1的行
    df_courses = pd.read_csv(course_csv_path)
    df_courses = df_courses[df_courses.iloc[:, -1] == 1]
    df_courses['rooms'] = df_courses['rooms'].apply(lambda x: room_names[x-1])
    df_courses['courses'] = df_courses['courses'].apply(lambda x: course_names[x-1])
    
    # 读取Workshop CSV文件，并只保留标记为1的行
    df_workshops = pd.read_csv(workshop_csv_path)
    df_workshops = df_workshops[df_workshops.iloc[:, -1] == 1]
    df_workshops['rooms'] = df_workshops['rooms'].apply(lambda x: room_names[x-1])
    df_workshops['courses'] = df_workshops.apply(lambda row: f"{course_names[row['courses']-1]} (Group {row['groups']})", axis=1)
    
    # 合并过滤后的课程和Workshop信息
    df_combined = pd.concat([df_courses.drop(df_courses.columns[-1], axis=1), df_workshops.drop(df_workshops.columns[-1], axis=1)], ignore_index=True)
    
    # 为每个房间生成课程表
    for room in df_combined['rooms'].unique():
        room_df = df_combined[df_combined['rooms'] == room]
        
        # 初始化空的DataFrame来存储当前房间的课程安排
        room_schedule = pd.DataFrame(index=pd.MultiIndex.from_product([room_df['weeks'].unique(), room_df['days'].unique()],
                                                                       names=['Week', 'Day']),
                                      columns=room_df['slots'].unique()).sort_index().sort_index(axis=1)
        
        # 填充课程表
        for _, row in room_df.iterrows():
            current_slot = room_schedule.at[(row['weeks'], row['days']), row['slots']]
            new_entry = row['courses']
            if pd.isna(current_slot):
                room_schedule.at[(row['weeks'], row['days']), row['slots']] = new_entry
            else:
                room_schedule.at[(row['weeks'], row['days']), row['slots']] = f"{current_slot}; {new_entry}"
        
        # 生成Excel文件名
        excel_file_name = f'{room}_schedule_filtered_S2.xlsx'
        
        # 创建一个新的Excel工作簿和工作表
        wb = Workbook()
        ws = wb.active
        ws.title = "Schedule"
        
        # 将DataFrame转换为工作表中的行
        for r_idx, row in enumerate(dataframe_to_rows(room_schedule, index=True, header=True), 1):
            for c_idx, value in enumerate(row, 1):
                ws.cell(row=r_idx, column=c_idx, value=value)
        
        # 调整列宽
        for column_cells in ws.columns:
            length = max(len(str(cell.value)) for cell in column_cells)
            ws.column_dimensions[get_column_letter(column_cells[0].column)].width = length
        
        # 保存工作簿
        wb.save(filename=excel_file_name)
        
        print(f'Saving schedul of {room}：{excel_file_name}')


# 使用函数生成带有课程名称的课程表
course_names = [
    "Advanced Methods of Applied Mathematics",
    "Algebraic Geometry",
    "Algebraic Topology",
    "Applied Dynamical Systems",
    "Bayesian Data Analysis",
    "Biomedical Data Science",
    "Calculus and its Applications",
    "Calculus and its Applications",  # Note: Repeated
    "Computing and Numerics",
    "Credit Scoring",
    "Data Assimilation",
    "Engineering Mathematics 1b",
    "Financial Mathematics",
    "Financial Risk Theory",
    "Fourier Analysis",
    "Fundamentals of Pure Mathematics",
    "Fundamentals of Pure Mathematics",  # Note: Repeated
    "Further Complex Variables",
    "Galois Theory",
    "Geometry of General Relativity",
    "Honours Algebra",
    "Honours Complex Variables",
    "Incomplete Data Analysis",
    "Integer and Combinatorial Optimization",
    "Interactions in Algebra, Geometry, and Topology",
    "Introduction to Number Theory",
    "Large Scale Optimization for Data Science",
    "Linear Programming, Modelling and Solution",
    "Machine Learning in Python",
    "Mathematics for the Natural Sciences 1b",
    "Mathematics in Action A",
    "Mathematics in Action B",
    "Metric Spaces",
    "Multivariate Data Analysis",
    "Numerical Methods for Data",
    "Numerical Ordinary Differential Equations and Applications",
    "Numerical Partial Differential Equations",
    "Numerical Probability and Monte Carlo",
    "Operational Research in the Energy Industry",
    "Optimization Methods in Finance",
    "Probability and Statistics",
    "Proofs and Problem Solving",
    "Proofs and Problem Solving",  # Note: Repeated
    "Research Skills for Financial Mathematics",
    "Risk and Logistics",
    "Risk-Neutral Asset Pricing",
    "Simulation",
    "Statistical Computing",
    "Statistical Research Skills",
    "Statistics (Year 2)",
    "Stochastic Control and Dynamic Asset Allocation",
    "Targeted Causal Learning",
    "Time Series",
    "Topics in Applied Operational Research",
    "Topics in Mathematical Biology",
    "Topics in Mathematical Physics A",
    "Topics in Mathematical Physics B"
]

room_names = [
    "JCMB_1501", "JCMB_2901", "JCMB_5326", "JCMB_5327", "JCMB_5328", "JCMB_6206",
    "JCMB_Lecture Theatre A", "JCMB_Lecture Theatre B", "JCMB_Lecture Theatre C",
    "NUC_1.14 - Oak Lecture Theatre", "NUC_B.01 - Alder Lecture Theatre", "NUC_1.15 - Larch Lecture Theatre"
]
generate_schedules('Course_S2.csv', 'Workshop_S2_update.csv',course_names, room_names)
```
